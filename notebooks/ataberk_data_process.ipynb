# DUD-E

```python
import pandas as pd

data_folder_path = "../data/ataberk/dude"

# mol = "abl1"
# mol = "reni"
mol = "thb"


actives_path = data_folder_path + f"/vectors/{mol}_actives.csv"
decoys_path = data_folder_path + f"/vectors/{mol}_decoys.csv"

actives_df = pd.read_csv(actives_path)
decoys_df = pd.read_csv(decoys_path)

actives_df['label'] = 1
decoys_df['label'] = 0


data_df = pd.concat([actives_df, decoys_df], ignore_index=True)
data_df = data_df.sample(frac=1).reset_index(drop=True)

data_df = data_df.sample(frac=1).reset_index(drop=True)

data_df.rename(columns={"smiles": "text", "label": "labels", "descriptors": "vectors"}, inplace=True)
data_df = data_df.reindex(columns=["text", "labels", "vectors"])

data_df[["text", "labels"]].to_csv(data_folder_path + f"/smiles/{mol}.csv", index=False)
data_df[["text", "vectors"]].to_csv(data_folder_path + f"/vectors/{mol}.csv", index=False)
```
